## Definição do Spider

```python
class ReposSpider(scrapy.Spider):
    name = "repos"
    download_delay = 1.0
    autothrottle_enabled = True

    def start_requests(self):
        urls = [f'https://github.com/search?p={page}&q=blockchain+stars%3A%3E%3D50&type=Repositories' for page in range(1, 2)]

        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        for repo in response.css('li.repo-list-item'):
            yield {
                'title': "".join(repo.xpath('.//a[@class="v-align-middle"]//text()').extract()),
                'description': "".join(repo.xpath('.//p[@class="mb-1"]//text()').extract()).strip(),
                'topics': [topic.strip() for topic in repo.css('a.topic-tag::text').extract()],
            }

process = CrawlerProcess(settings={
    "FEEDS": {
        "data-spider.csv": {"format": "csv"},
    },
})

process.crawl(ReposSpider)
process.start() # the script will block here until the crawling is finished
```
